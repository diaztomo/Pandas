## Persiapan Dataframe

-Kita akan membuat suatu dataframe terdiri dari 5 kolom dan 5 baris dengan nama kolom berurut A,B,C,D,E

-Entry dari tiap kolom merupakan bilangan integer acak dari method np.random.randint()

-Paramater dari fungsi random adalah np.random.randint(a,b,size = (baris,kolom))

-a = nilai interval terkecil dari suatu nilai acak

-b = nilai interval terbesar dari suatu nilai acak

-size=(n_baris,n_kolom) = ukuran array yang akan digenerate sejumlah n_baris dan n_kolom

```python
n_rows = 5 
n_columns = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1,10,size = (n_rows,n_columns)),
                  columns = cols)

df
```

```python
tuple("ABCDE")
```

## Menambahkan Prefix pada kolom

Prefik atau awalan adalah kata awalan sebelum kata dasar.

Kita akan menggunakan method DataFrame.add_prefix(INPUT) untuk memberi awalan pada tiap kolom. input merupakan string yang dimasukan pada INPUT

```python
df.add_prefix("Kolom_")
```

## Menambahkan Suffix pada kolom

Suffix atau akhiran adalah kata yang ditaruh setelah kata dasar

Kita akan menggunakan method DataFrame.add_suffix(INPUT) untuk memberi akhiran pada tiap kolom. input merupakan string yang dimasukan pada INPUT

```python
df.add_suffix("_Field")
```

Buatlah Dataframe berisi 20 baris dan 7 kolom dari kolom AA,AB,AC,AD,AE,AF,AG. dimana isi dari entry merupakan bilang acak menyebar normal dari 2 sampai 100

a. menambahkan prefix "kolom_"
b. menambahkan suffic "_field"

```python
n_rows = 20
n_cols = 7
cols = tuple("ABCDEFG")

df = pd.DataFrame(np.abs(np.random.randn(n_rows, n_cols))+2, columns = cols)

df = df.add_prefix("A")
df
```

```python
df.add_prefix("Kolom_")
```
